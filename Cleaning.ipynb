```python
cols = ['DATE_OCC', 'TIME_OCC', 'AREA_NAME', 'Crm_Cd_Desc', 'Vict_Age', 'Vict_Sex', 'Vict_Descent', 'Weapon_Desc', 'Status_Desc', 'LAT', 'LON', 'crm_cat', 'total_crm', 'VAge_Grp', 'year']
```

```python
df = df.loc[:, cols]
```

```python
def change_vict_desc(x):
    if x == 'A':
        return 'Other Asian'
    elif x == 'B':
        return 'Black'
    elif x == 'C':
        return 'Chinese'    
    elif x == 'D':
        return 'Cambodian'    
    elif x == 'F':
        return 'Filipino'
    elif x == 'G':
        return 'Guamanian'
    elif x == 'H':
        return 'Hispanic/Latin/Mexican'
    elif x == 'I':
        return 'American Indian/Alaskan Native'
    elif x == 'J':
        return 'Japanese'
    elif x == 'K':
        return 'Korean'
    elif x == 'L':
        return 'Laotian'
    elif x == 'O':
        return 'Other'
    elif x == 'p':
        return 'Pacific Islander'
    elif x == 'S':
        return 'Samoan'
    elif x == 'U':
        return 'Hawaiian'
    elif x == 'V':
        return 'Vietnamese'
    elif x == 'W':
        return 'White'
    elif x == 'X':
        return 'Unknown'
    elif x == 'Z':
        return 'Asian Indian'
    else:
        return 'Unknown'
```

```python
df['Vict_Descent'] = df['Vict_Descent'].apply(change_vict_desc)
```
